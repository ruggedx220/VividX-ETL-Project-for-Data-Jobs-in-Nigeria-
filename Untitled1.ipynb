```python
import pandas as pd

# Read the CSV file with a specified encoding
csv = ('JobsLin_data.csv')
df_1 = pd.read_csv('JobsLin_data.csv', encoding='latin1')


# Define the columns you want to add
new_columns = ["Company", "Location", "Salary"]



# Add the new columns to the DataFrame with empty values
for column in new_columns:
    df_1[column] = ''

# Write the DataFrame back to the CSV file
df_1.to_csv(csv, index=False)
# Display the DataFrame
df_1
```

```python
column_rename_mapping = {
    'title': 'Title',
    'description': 'Description',
    'requirement': 'Requirement',
    'description': 'Description',
    'job_type': 'Job_type',
    # Add more mappings as needed
}

# Use the rename() method to rename columns in the DataFrame
df_1.rename(columns=column_rename_mapping, inplace=True)
df_1
```
